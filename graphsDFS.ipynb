```python
#Array of Edges
#Diected graphs

n=8
A=[[0,1],[1,2],[0,3],[3,4],[3,6],[3,7],[4,2],[4,5],[5,2]]
print(A)
```

```python
from collections import defaultdict

D = defaultdict(list)

for u,v in A:
    D[u].append(v)
    #if its an undirected graph then uncomment the line below
    #D[v].append(u)

D
```

```python
#Recursive dfs - O(V+E) V is number of nodes E is number of edges 
source = 0

def dfs_recursive(node):
    print(node)
    for nei_node in D[node]:
        if nei_node not in seen:
            seen.add(nei_node)
            dfs_recursive(nei_node)

seen = set()

seen.add(source)

dfs_recursive(source)
```

```python
#Iterative DFS with a stack  - O(V+E)

source = 0

seen = set()

seen.add(source)
stack = [source]

while stack:
    node = stack.pop()
    print(node)
    for nei_node in D[node]:
        if nei_node not in seen:
            seen.add(nei_node)
            stack.append(nei_node)
```
