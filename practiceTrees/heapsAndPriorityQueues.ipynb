```python
import heapq

A = [-4, 3, 1, 0, 2, 5, 10, 8, 12, 9]

print(A)
```

```python
heapq.heapify(A)

print(A)
```

```python
heapq.heappush(A,4)

print(A)
```

```python
minn = heapq.heappop(A)
print(minn, A)
```

```python
#Heap Sort

def heapSort(arr):
    heapq.heapify(arr)
    n = len(arr)
    new_list = [0]*n
    
    for i in range(n):
        m = heapq.heappop(arr)
        new_list[i] = m
        
    return new_list

heapSort(A)
```

```python
#Heap push and pop
heapq.heappushpop(A,99)
print(A)
```

```python
#Building a Max Heap

B = [-4, 3, 1, 0, 2, 5, 10, 8, 12, 9]
n = len(B)

for i in range(n):
    B[i]*=-1
    
print(B)

#Apply min heap

heapq.heapify(B)

print(B)

#get largest

largest = -1*heapq.heappop(B)

print(largest, B)


#inserting into a maxheap
valToInsert = 7
heapq.heappush(B, -1*valToInsert)
print(B)

```

```python
#BUilding a heap from scratch

C = [-5, 4, 2, 1, 7, 0, 3]

heap = []

for x in C:
    heapq.heappush(heap, x)
    print(heap, len(heap))
```

```python
#putting tupples of numbers on to the heap
from collections import Counter

D = [5, 4, 3, 5, 4, 3, 5, 4, 3, 5, 5, 4, 4, 8]

counter = Counter(D)

print(counter)

heap = []

for k,v in counter.items():
    print(k,v)
    heapq.heappush(heap, (v,k))
    
print(heap)
    
```
