```python
class TreeNode:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left 
        self.right = right
        
    def __str__(self):
        return str(self.value)
```

```python
A = TreeNode(1)
B = TreeNode(2)
C = TreeNode(3)
D = TreeNode(4)
E = TreeNode(5)
F = TreeNode(10)

A.left = B
A.right = C
B.left = D
B.right = E
C.left = F

print(A)
```

### DFS Search

```python
def dfsSearch(node, target):
    if not node:
        return False
    print(node)
    if node.value == target:
        return True
    return (dfsSearch(node.left, target) or dfsSearch(node.right, target))
    
print(dfsSearch(A,10))
```

### BFS Search

```python
from collections import deque

def bfsSearch(node, target) -> bool:
    q = deque()
    
    q.append(node)
    
    while q:
        node = q.popleft()
        print(node)
        
        if node.value == target:
            return True
        
        if node.left:
            q.append(node.left)
        if node.right:
            q.append(node.right)
    
    return False

bfsSearch(A, 10)
```
