```python
class TreeNode:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left 
        self.right = right
        
    def __str__(self):
        return str(self.value)
            
```

```python
A = TreeNode(1)
B = TreeNode(2)
C = TreeNode(3)
D = TreeNode(4)
E = TreeNode(5)
F = TreeNode(10)

A.left = B
A.right = C
B.left = D
B.right = E
C.left = F

print(A)
```

# DFS Traversals

### InOrder Traversal (Recursive)

```python
def inOrder(node):
    if not node:
        return
    
    inOrder(node.left)
    print(node)
    inOrder(node.right)

inOrder(A)
```

### PreOrder Traversal (Recursive)

```python
def preOrder(node):
    if not node:
        return
    
    print(node)
    preOrder(node.left)
    preOrder(node.right)
    
preOrder(A)
```

### PostOrder Traversal (Recursive)

```python
def postOrder(node):
    if not node:
        return
    postOrder(node.left)
    postOrder(node.right)
    print(node)
    
postOrder(A)
```

### PreOrder Traversal (With Stack)

```python
def preOrderItr(node):
    stk = [node]
    while stk:
        node = stk.pop()
        print(node)
        if node.right:
            stk.append(node.right)
        if node.left:
            stk.append(node.left)
            
preOrderItr(A)
        
```

# BFS Traversals

```python
from collections import deque

def lvlOrder(node):
    q = deque()
    q.append(node)
    while q:
        node = q.popleft()
        print(node)
        if node.left: q.append(node.left)
        if node.right: q.append(node.right)
        
lvlOrder(A)
```
