```python
#Array of Edges
#Diected graphs

n=8
A=[[0,1],[1,2],[0,3],[3,4],[3,6],[3,7],[4,2],[4,5],[5,2]]
print(A)
```

```python
#Adjency Matrix

M=[]

for i in range(n):
    M.append([0]*n)
    
for u,v in A:
    M[u][v] = 1
    #if its an undirected graph then we uncomment the line below
    M[v][u] = 1
    
M

```

```python
# Array of Edges -> Adjacency List
from collections import defaultdict

D = defaultdict(list)

for u,v in A:
    D[u].append(v)
    #if its an undirected graph then uncomment the line below
    D[v].append(u)

D
```
